```python
import pandas as pd
```

```python
df = pd.read_csv('raw_data/sales.csv', encoding='unicode_escape')
df.head()
```

```python
df.info()
```

#### 날짜 데이터 형변환

```python
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%m%d%Y %H%m%s', infer_datetime_format=True)
```

```python
df.head()
```

```python
df.info()
```

#### 결측 데이터 제거

```python
df.isnull().sum()
```

```python
df = df[~df['CustomerID'].isnull() & ~df['Description'].isnull()].copy()
```

```python
df = df[df['Quantity'] > 0].copy()
```

```python
df
```

#### 고객 아이디 데이터 타입 변환

```python
df['CustomerID'] = df['CustomerID'].astype('int').astype('object')
```

```python
df
```

```python
df.info()
```

```python
df.to_csv('sales_data.csv', index=False)
```
